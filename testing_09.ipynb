```python
import requests
import pandas as pd
from time import sleep
```

```python
def fetch_reports(org_code="80000031", start_page=1, end_page=500, page_size=20, sleep_sec=0.3):
    url = "https://reportapi.eastmoney.com/report/list"
    headers = {"User-Agent": "Mozilla/5.0"}
    all_data = []

    for p in range(start_page, end_page + 1):
        params = {
            "orgCode": org_code,
            "pageNo": p,
            "pageSize": page_size,
            "industryCode": "*",
            "industry": "*",
            "rating": "*",
            "ratingChange": "*",
            "qType": 0,
            "code": "",
            "beginTime": "",
            "endTime": "",
            "_": int(__import__('time').time() * 1000)
        }

        try:
            r = requests.get(url, headers=headers, params=params, timeout=15)
            r.raise_for_status()
            data = r.json().get("data", [])
        except Exception as e:
            print(f"第 {p} 页请求失败: {e}")
            continue

        if not data:
            print(f"第 {p} 页无数据，提前结束")
            break

        all_data.extend(data)
        print(f"已获取第 {p} 页，累计 {len(all_data)} 条")
        sleep(sleep_sec)

    return pd.DataFrame(all_data)
```
